```python
import re
import spacy
from pdfminer.high_level import extract_text

# Load spaCy model
nlp = spacy.load("en_core_web_sm")

# Extract text from PDF resume
def extract_resume_text(file_path):
    return extract_text(file_path)

# Extract email using regex
def extract_email(text):
    email = re.findall(r"[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,4}", text)
    return email[0] if email else None

# Extract phone number using regex
def extract_phone(text):
    phone = re.findall(r"\+?\d[\d\s-]{8,13}\d", text)
    return phone[0] if phone else None

# Extract name using spaCy NER
def extract_name(text):
    doc = nlp(text)
    for ent in doc.ents:
        if ent.label_ == "PERSON":
            return ent.text
    return None

# Extract skills using keyword matching
def extract_skills(text):
    skills_list = [
        "python", "java", "c++", "sql", "machine learning",
        "deep learning", "hadoop", "spark", "nlp"
    ]

    text = text.lower()
    found_skills = [skill for skill in skills_list if skill in text]
    return found_skills

# Main parser function
def parse_resume(file_path):
    text = extract_resume_text(file_path)

    data = {
        "Name": extract_name(text),
        "Email": extract_email(text),
        "Phone": extract_phone(text),
        "Skills": extract_skills(text)
    }

    return data

# Run the parser
if __name__ == "__main__":
    # To resolve FileNotFoundError, you need to either:
    # 1. Upload a 'resume.pdf' file to your Colab environment (e.g., via the file explorer icon on the left).
    # 2. Change 'file_path' to an existing PDF file path.
    # 3. Create a dummy 'resume.pdf' for testing (e.g., using a library like PyPDF2 or reportlab).
    file_path = "resume.pdf"  # Replace with your resume file path
    result = parse_resume(file_path)

    print("Parsed Resume Data:\n")
    for key, value in result.items():
        print(f"{key}: {value}")
```
