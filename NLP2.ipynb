```python
import re

# Define ELIZA rules: (pattern, response)
rules = [
    (r'.*\bI am unhappy\b.*',
     "DO YOU THINK COMING HERE WILL HELP YOU NOT TO BE UNHAPPY"),

    (r'.*\bI need (.*)\b.*',
     "WHAT WOULD IT MEAN TO YOU IF YOU GOT \\1"),

    (r'.*\bmother\b.*',
     "TELL ME MORE ABOUT YOUR FAMILY"),

    (r'.*\bMy mother takes care of me\b.*',
     "WHO ELSE IN YOUR FAMILY TAKES CARE OF YOU"),

    (r'.*\bfather\b.*',
     "YOUR FATHER"),

    (r'.*\bYou are like my father\b.*',
     "WHAT MAKES YOU THINK I AM LIKE YOUR FATHER"),
]

def eliza_response(user_input):
    user_input = user_input.strip()

    for pattern, response in rules:
        match = re.match(pattern, user_input, re.IGNORECASE)
        if match:
            # Replace captured groups if any
            return re.sub(r'\\(\d+)',
                          lambda m: match.group(int(m.group(1))),
                          response)

    return "PLEASE TELL ME MORE."

# Chat loop
print("ELIZA: HELLO. HOW CAN I HELP YOU?")
while True:
    user = input("User: ")
    if user.lower() in ["quit", "exit", "bye"]:
        print("ELIZA: GOODBYE. TAKE CARE.")
        break
    print("ELIZA:", eliza_response(user))
```
